```python
# Answer 1

import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

L = 200
lst= []
for x in range(0,L):
  lst.append(np.random.randint(-10,10))

newlst = []

for count in range(1, len(lst)+1):
  avg = 0
  for i in range(count, 0, -1):
      avg  += lst[i-1]

  newlst.append(round(avg/len(lst),2))

print(lst)
print(newlst)

plt.plot(newlst)
```

```python
#Answer 2

import numpy as np
S0 = 100 
K=110
sigma = 0.05
deltat = 0.01
N=np.zeros(1000) 

for i in range(0,N.size):
  y = S0
  for j in range(1, N.size):
      y = y * np.exp(np.random.choice([-1,1]) * sigma * np.sqrt(deltat))
  N[i]=y

C=np.maximum(0,N-K)
P=np.maximum(0,K-N)
print("The call price is $", round(C.mean(),4))
print("The put price is $", round(P.mean(),4))
```
